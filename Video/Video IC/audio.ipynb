```python
from IPython.display import Audio
%cd "Video/Video IC/audio"
```

```python
def generate_audio(text, filename, voice='pt-BR-ThalitaNeural'):
    
    command = f'! edge-tts --voice {voice} --text "{text}" --write-media {filename}'
    get_ipython().system(command)
    
    return Audio(filename)
```

```python
generate_audio("Neste vídeo você irá aprender os conceitos básicos por trás da computação quântica e seus benefícios sobre a computação clássica para resolução de problemas complexos.", "1.mp3")
```

```python
a = "A computação quântica, apesar de funcionalmente diferente da clássica, possui elementos em comum, o que permite traçar algumas comparações entre ambos."


generate_audio(f"{a}", "2.mp3")
```

```python
a = "em específico, os bits clássicos, blocos fundamentais da computação clássica, são análogos aos bits quânticos, ou kiubites, que são a base da computação quântica." 

generate_audio(f"{a}", "3.mp3")
```

```python
a = "Os bits clássicos são compostos por dois valores. 0 e 1. Por outro lado, os kiubites são representados pelos estados kete 0 e kete 1. Apresentados aqui na notação de Dirac."


generate_audio(f"{a}", "4.mp3")
```

```python
a = "Enquanto os bits clássicos podem ocupar dois possíveis valores, os kiubites podem ser representados por uma combinação linear de estados, chamado de superposição."


generate_audio(f"{a}", "5.mp3")
```

```python
a = "Sendo assim, os estados quânticos dos kiubites podem ser considerados como um vetor em um espaço vetorial complexo bidimensional, onde os estados kete 0 e kete 1, naturalmente ortogonais, formam a base do espaço vetorial, chamada de base computacional."


generate_audio(f"{a}", "6.mp3")
```

```python
a = "Uma das propriedades mais importantes de estados quânticos, é de que não é possível medir o estado psi de um kiubite sem alterá-lo. Isso nos diz que após realizar uma medida de um estado psi de um kiubite, o estado do kiubite será alterado para um dos estados da base computacional, kete 0 ou kete 1. A probabilidade de medir o estado kete 0 é de módulo de alpha ao quadrado, e a probabilidade de medir o estado kete 1 é de módulo de beta ao quadrado. Onde a soma dessas probabilidades tem que ser igual a 1."


generate_audio(f"{a}", "7.mp3")
```

```python
a = "Sendo assim, o estado de um kiubite deve pertencer à circunferencia unitária no plano complexo, de modo a respeitar as probabilidades alpha ao quadrado ou probabilidade de obter 0 e beta ao quadrado ou probabilidade de obter 1."


generate_audio(f"{a}", "8.mp3")
```

```python
voices = ["pt-BR-ThalitaNeural", "pt-BR-FranciscaNeural", "pt-BR-AntonioNeural"]

a = "Olá! Hoje vamos explorar o mundo da computação quântica aplicada a materiais."

for voice in voices:
    
    generate_audio(f"{a}", f"audio_1_{voice}.mp3", voice)
```

```python
a = "Agora apresentaremos uma notação diistinta para lidar com os estados (Psi) dos kiubites. Como os coeficientes alpha e beta são números complexos, podemos utilizar a forma polar para representálos, dada pela seguinte expressão. Onde z é um número complexo, r é o módulo de z e theta é o argumento de z."

generate_audio(f"{a}", "9.mp3")
```

```python
a = "Agora apresentaremos uma notação diistinta para lidar com os estados (Psi) dos kiubites. Como os coeficientes alpha e beta são números complexos, podemos utilizar a forma polar para representálos, dada pela seguinte expressão. Onde z é um número complexo, a e b são numeros reais. Ou ainda, utilizando a notação de Êuler, onde r é o módulo de z ,e theta é o argumento de z."

generate_audio(f"{a}", "10.mp3")
```

```python
a = "Agora, realizando algumas manipulações algébricas, podemos reescrever o estado de um kiubite em termos dos raios e respectivos ângulos theta, apresentando assim um coeficiente global multiplicando a superposição dos estados kete 0 e kete 1, chamado de fase global."

generate_audio(f"{a}", "11.mp3")
```

```python
a = "A seguir, vamos provar que a fase global de um kiubite é totalmente irrelevante, dependendo apenas da fase relativa, que corresponde a diferença dos ângulos theta. Sendo assim, introduzimos um novo estado, kete Psi linha, que corresponde ao mesmo estado que kete psi, mas sem a fase global."

generate_audio(f"{a}", "12.mp3")
```

```python
a = "Agora, para comparar ambos os estados e provar que são totalmente equivalentes, faremos uma análise do valor esperado para um dado operador, que é uma medida de uma propriedade física do sistema. Para isso, utilizaremos um operador genérico, o operador Ó."

generate_audio(f"{a}", "13.mp3")
```

```python
a = "Realizando a substituição de kete psi linha por kete psi, multiplicado pela fase global, obtemos que o valor esperado para o operador Ó é o mesmo para ambos os estados, o que nos diz que a fase global de um kiubite é totalmente irrelevante para a computação quântica, dependendo apenas da fase relativa entre alpha e beta, que corresponde a theta 0 e theta 1."

generate_audio(f"{a}", "14.mp3")
```

```python
a = "Sendo assim, de agora em diante vamos ignorar a fase global dos kiubites, focando apenas na fase relativa entre os estados kete 0 e kete 1. Além disso, vamos determinar uma nova notação para a fase relativa, délta, que corresponde a essa diferença de fases."

generate_audio(f"{a}", "15.mp3")
```
